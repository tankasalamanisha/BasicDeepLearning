### Human Activity Recognition

Multivariate TimeSeries Analysis.

Includes sensor data from accelerometers and gyroscopes.

6-Class classifier.

```python
import numpy as np
import pandas as pd

# getting features:

with open('../data/features.txt') as f:
    features = [line.split()[1] for line in f.readlines()]
    
print(f"No. of features:{len(features)}")
```

```python
len(set(features))
```

```python
# Reading the training data

X_train = pd.read_csv('../data/train/X_train.txt',delim_whitespace=True, header=None)
```

```python
X_train
```

```python
X_train.columns = features
```

```python
X_train
```

```python
X_train['subject'] = pd.read_csv('../data/train/subject_train.txt',header=None, squeeze=True)
```

```python
X_train
```

```python
y_train = pd.read_csv("../data/train/y_train.txt",header=None,squeeze=True)
```

```python
y_train.columns= ["Activity"]
```

```python
y_train
```

```python
activity_map = {
    1: "WALKING",
    2:'WALKING_UPSTAIRS',
    3:'WALKING_DOWNSTAIRS',
    4:'SITTING', 
    5:'STANDING',
    6:'LAYING'
}
```

```python
y_train_labels=y_train.map(activity_map)
```

```python
train = X_train
train["Activity"] = y_train
train["ActivityName"] = y_train_labels
```

```python
train.sample()
```

```python
# Reading test data
X_test = pd.read_csv("../data/test/X_test.txt",delim_whitespace=True,header=None)
```

```python
X_test.columns = features
```

```python
X_test["subject"] = pd.read_csv("../data/test/subject_test.txt",header=None, squeeze=True)
```

```python
y_test = pd.read_csv("../data/test/y_test.txt",header=None, squeeze=True)
```

```python
y_test_labels = y_test.map(activity_map)
```

```python
test = X_test
test["Activity"] = y_test
test["ActivityName"] = y_test_labels
```

```python
test.sample()
```

```python
## Data Imbalance
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['font.family'] = 'Dejavu Sans'

plt.figure(figsize=(16,8))
plt.title('Data provided by each user', fontsize=20)
sns.countplot(x='subject',hue='ActivityName', data = train)
plt.show()
```
